Removing the prefix "Iris" from the column Class(Species)

```python
df['Class(Species)']=df['Class(Species)'].str.replace("Iris-","")
```

Dividing the data into training and test data in 80 to 20 ratio

```python
X_train,X_test,Y_train,Y_test = train_test_split(df.iloc[:,:4],df['Class(Species)'],test_size=0.2,random_state=2)
```

```python
lr=LogisticRegression()
```

Training the model on the divided dataset

```python
lr.fit(X_train,Y_train)
```

Storing the target variable in the variable pred

```python
pred=lr.predict(X_test)
```

```python
print(pred)
```

Checking the accuracy of the model on test data

```python
(lr.score(X_test,Y_test))*100
```

This accuracy score shows the prediction accuracy of the model.
